```python
import matplotlib.pyplot as plt
#Sample data
students = ["A","B","C","D","E"]
marks = [85,70,90,60,75]

plt.bar(students,marks,color="skyblue")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Student Performance")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#Sample data
tips = sns.load_dataset("tips")
sns.boxplot(x="day",y="total_bill",data=tips)
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.title("Box Plot of Total Bill by Day")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data={"Month":["Jan","Feb","Mar","Apr","May"],"Sales":[1000,1200,900,1500,1100]}
df=pd.DataFrame(data)
df.plot(x="Month",y="Sales",kind="bar",color="skyblue")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales")
```

```python
days=["mon","tues","wed","thurs","fri","sat","sun"]
temperature=[30,32,31,29,33,35,34]
plt.plot(days,temperature,marker='s',linestyle='--',color='red')
plt.title("weekly temperature trend")
plt.xlabel("days")
plt.ylabel("temperature")
plt.show()
```

```python
import matplotlib.pyplot as plt
months=["jan","feb","mar","apr","may"]
sales=[1000,1200,900,1500,1100]
plt.bar(months,sales,color="green")
plt.title("monthly sales")
plt.xlabel("months")
plt.ylabel("sales")
plt.grid(axis='y')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
tips = sns.load_dataset("tips")

# Barplot
plt.figure(figsize=(6, 4))
sns.barplot(x="day", y="total_bill", data=tips, palette="muted")
plt.title("Barplot Example")
plt.show()

# Boxplot
plt.figure(figsize=(6, 4))
sns.boxplot(x="day", y="total_bill", data=tips, palette="Set2")
plt.title("Boxplot Example")
plt.show()

# Violinplot
plt.figure(figsize=(6, 4))
sns.violinplot(x="day", y="total_bill", data=tips, palette="Set3")
plt.title("Violinplot Example")
plt.show()

# Countplot
plt.figure(figsize=(6, 4))
sns.countplot(x="day", data=tips, palette="coolwarm")
plt.title("Countplot Example")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset("iris")
sns.pairplot(df,hue="species",diag_kind="kde",palette="set2")
plt.suptitle("pairplot example-iris dataset",y=1.02)
plt.show()
```
